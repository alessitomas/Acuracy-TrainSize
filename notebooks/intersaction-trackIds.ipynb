```python
import pandas as pd 
```

```python
fma = pd.read_csv("../data/fma/fmaManipulated.csv")
```

```python
fma.head()
```

```python
fma_small = fma[fma['subset'] == 'small']
fma_medium = fma[fma['subset'] == 'medium']
fma_large = fma[fma['subset'] == 'large']
```

```python
fma_small_handcrafted = pd.read_csv('../data/fma/fma-small-handcrafted.csv')
fma_medium_handcrafted = pd.read_csv('../data/fma/fma-medium-handcrafted.csv')
fma_large_handcrafted = pd.read_csv('../data/fma/fma-large-handcrafted.csv')
```

```python
fma_small
```

```python
intersection = {}

fma_small_intersaction = fma_small[fma_small['track_id'].isin(fma_small_handcrafted['track_id'])]
fma_medium_intersaction = fma_medium[fma_medium['track_id'].isin(fma_medium_handcrafted['track_id'])]
fma_large_intersaction = fma_large[fma_large['track_id'].isin(fma_large_handcrafted['track_id'])]
```

```python
intersection = {}
intersection['fma-small'] = fma_small_intersaction['track_id'].tolist()
intersection['fma-medium'] = fma_medium_intersaction['track_id'].tolist()
intersection['fma-large'] = fma_large_intersaction['track_id'].tolist()
```

```python
len(intersection['fma-small']), len(intersection['fma-medium']), len(intersection['fma-large'])
```

```python
intersection['fma-small']
```

```python
intersection['fma-medium']
```

```python
import json

with open('../data/intersection-trackIds.json', 'w') as file:
    json.dump(intersection,file)
```

```python
def readStorage(storageJsonPath):
    with open(storageJsonPath, 'r') as f:
        storage = json.load(f)
        return storage
```

```python
intersaction_tack_ids = readStorage('../data/intersection-trackIds.json')
```

```python
teste = intersaction_tack_ids['fma-small']
```
