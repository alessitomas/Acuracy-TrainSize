```python
FMA_tracks = pd.read_csv("../data/fma/tracks.csv", index_col=0)
FMA_tracks.columns = FMA_tracks.iloc[0]
```

```python
track_info = FMA_tracks[["subset","genre_top"]]
track_info.head()
```

```python
track_info.columns.rename('',inplace=True)
track_info = track_info.iloc[2:,:]
track_info = track_info[track_info['genre_top'].isnull() == False] 
track_info.head()
```

```python
track_info['track_id'] = track_info.index
```

```python
track_info
```

```python
FAM_features = pd.read_csv("../data/hand-crafted/FMA_medium_handcrafted_features.csv")
```

```python
FAM_features.head()
```

```python
# Concatenate the DataFrames based on the 'track_id' column
merged_df = pd.merge(FAM_features, track_info, on='track_id', how='inner')
```
