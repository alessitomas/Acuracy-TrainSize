```python
import json

def readStorage(storageJsonPath):
    with open(storageJsonPath, 'r') as f:
        storage = json.load(f)
        return storage
```

```python
storageJsonPath = '../data/newm.json'
metricsJson = readStorage(storageJsonPath)
```

```python
metricsGTZAN = metricsJson['GTZAN']
metricsFMAsmall = metricsJson['FMA-small']
metricsFMAmedium = metricsJson['FMA-medium']
metricsFMAlarge = metricsJson['FMA-large']
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6),dpi=1000)

# plot accuracy score GTZAN
plt.plot(metricsGTZAN['n_train_'], metricsGTZAN['acc_mean'], markersize=4, label='GTZAN', color='#FF7F0E')
plt.errorbar(metricsGTZAN['n_train_'], metricsGTZAN['acc_mean'], yerr= [2*i for i  in metricsGTZAN['acc_std']], fmt='o', markersize=4, capsize=3, capthick=1, ecolor='black', color='black')

# plot accuracy score GTZAN overfit
plt.plot(metricsGTZAN['n_train_'], metricsGTZAN['acc_mean_overfit'], markersize=4, color='#FF7F0E', animated=True,linestyle=':')
plt.errorbar(metricsGTZAN['n_train_'], metricsGTZAN['acc_mean_overfit'], yerr=[2*i for i  in metricsGTZAN['acc_std_overfit']], fmt='o', markersize=4, capsize=3, capthick=1, ecolor='black', color='black')

# plot accuracy score FMA small
plt.plot(metricsFMAsmall['n_train_'], metricsFMAsmall['acc_mean'], markersize=4, label='FMA-small', color='red')
plt.errorbar(metricsFMAsmall['n_train_'], metricsFMAsmall['acc_mean'], yerr=[2*i for i  in metricsFMAsmall['acc_std']], fmt='o', markersize=4, capsize=3, capthick=1, ecolor='black', color='black')

# plot accuracy score FMA small overfit
plt.plot(metricsFMAsmall['n_train_'], metricsFMAsmall['acc_mean_overfit'], markersize=4, color='red', animated=True,linestyle=':')
plt.errorbar(metricsFMAsmall['n_train_'], metricsFMAsmall['acc_mean_overfit'], yerr=[2*i for i  in metricsFMAsmall['acc_std_overfit']] , fmt='o', markersize=4, capsize=3, capthick=1, ecolor='black', color='black')


# plot accuracy score FMA medium
plt.plot(metricsFMAmedium['n_train_'], metricsFMAmedium['acc_mean'], markersize=4, label='FMA-medium', color='blue')
plt.errorbar(metricsFMAmedium['n_train_'], metricsFMAmedium['acc_mean'], yerr=[2*i for i  in metricsFMAmedium['acc_std']], fmt='o', markersize=4, capsize=3, capthick=1, ecolor='black', color='black')

# plot accuracy score FMA medium
plt.plot(metricsFMAmedium['n_train_'], metricsFMAmedium['acc_mean_overfit'], markersize=4, color='blue', linestyle=':')
plt.errorbar(metricsFMAmedium['n_train_'], metricsFMAmedium['acc_mean_overfit'], yerr=[2*i for i  in metricsFMAmedium['acc_std_overfit']], fmt='o', markersize=4, capsize=3, capthick=1, ecolor='black', color='black')


# # plot accuracy score FMA large
# plt.plot(metricsFMAlarge['n_train_'], metricsFMAlarge['acc_mean'], markersize=4, label='FMA-large', color='purple')
# plt.errorbar(metricsFMAlarge['n_train_'], metricsFMAlarge['acc_mean'], yerr=metricsFMAlarge['acc_std'], fmt='o', markersize=4, capsize=3, capthick=1, ecolor='black', color='black')

# # plot accuracy score FMA large
# plt.plot(metricsFMAlarge['n_train_'], metricsFMAlarge['acc_mean_overfit'], markersize=4, color='purple', linestyle=':')
# plt.errorbar(metricsFMAlarge['n_train_'], metricsFMAlarge['acc_mean_overfit'], yerr=metricsFMAlarge['acc_std_overfit'], fmt='o', markersize=4, capsize=3, capthick=1, ecolor='black', color='black')




# set the range on y-axis
plt.ylim(0,1.1)
# Set x-axis to logarithmic scale
plt.xscale("log")

# set the title
plt.title("Different Datasets Claasification Accuracy vs Number of Training Examples")
# set the labels on x-axis and y-axis
plt.xlabel("Number of Traning Examples")
plt.ylabel("Accuracy Score")
# display the plots labels 
plt.legend()
# display the background grid
plt.grid(True, linestyle='--', linewidth=0.4, color='gray')
plt.show()
```
