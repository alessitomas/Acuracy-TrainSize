```python
import json

def readStorage(storageJsonPath):
    with open(storageJsonPath, 'r') as f:
        storage = json.load(f)
        return storage
```

```python
storageJsonPath = '../data/metrics.json'
metricsJson = readStorage(storageJsonPath)
```

```python
metricsGTZAN = metricsJson['gtzan']
metricsGTZAN_handcrafted = metricsJson['gtzan-handcrafted']
metricsFMAsmall = metricsJson['fma-small']
metricsFMAsmall_handcrafted = metricsJson['fma-small-handcrafted']
metricsFMAmedium = metricsJson['fma-medium']
metricsFMAmedium_handcrafted = metricsJson['fma-medium-handcrafted']
metricsFMAlarge = metricsJson['fma-large']
metricsFMAlarge_handcrafted = metricsJson['fma-large-handcrafted']
```

GTZAN reference: https://www.kaggle.com/datasets/andradaolteanu/gtzan-dataset-music-genre-classification/

FMA reference: https://www.kaggle.com/datasets/imsparsh/fma-free-music-archive-small-medium

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6),dpi=1000)

# plot accuracy score GTZAN
plt.plot(metricsGTZAN['n_train_'], metricsGTZAN['acc_mean'], markersize=4, label='GTZAN', color='red')
plt.errorbar(metricsGTZAN['n_train_'], metricsGTZAN['acc_mean'], yerr= [2*i for i  in metricsGTZAN['acc_std']], fmt='o', markersize=4, capsize=3, capthick=1, ecolor='black', color='black')

# plot accuracy score GTZAN overfit
plt.plot(metricsGTZAN['n_train_'], metricsGTZAN['acc_mean_overfit'], markersize=4, color='red', animated=True,linestyle=':')
plt.errorbar(metricsGTZAN['n_train_'], metricsGTZAN['acc_mean_overfit'], yerr=[2*i for i  in metricsGTZAN['acc_std_overfit']], fmt='o', markersize=4, capsize=3, capthick=1, ecolor='black', color='black')

#plot accuracy score GTZAN handcrafted
plt.plot(metricsGTZAN_handcrafted['n_train_'], metricsGTZAN_handcrafted['acc_mean'], markersize=4, color='crimson' ,label='GTZAN-handcrafted')
plt.errorbar(metricsGTZAN_handcrafted['n_train_'], metricsGTZAN_handcrafted['acc_mean'], yerr= [2*i for i  in metricsGTZAN_handcrafted['acc_std']], fmt='o', markersize=4, capsize=3, capthick=1, ecolor='black', color='black')

# plot accuracy score GTZAN handcrafted overfit
plt.plot(metricsGTZAN_handcrafted['n_train_'], metricsGTZAN_handcrafted['acc_mean_overfit'], markersize=4, color='crimson', animated=True,linestyle=':')
plt.errorbar(metricsGTZAN_handcrafted['n_train_'], metricsGTZAN_handcrafted['acc_mean_overfit'], yerr=[2*i for i  in metricsGTZAN_handcrafted['acc_std_overfit']], fmt='o', markersize=4, capsize=3, capthick=1, ecolor='black', color='black')


# plot accuracy score FMA small 
plt.plot(metricsFMAsmall['n_train_'], metricsFMAsmall['acc_mean'], markersize=4, label='FMA-small', color='blue')
plt.errorbar(metricsFMAsmall['n_train_'], metricsFMAsmall['acc_mean'], yerr=[2*i for i  in metricsFMAsmall['acc_std']], fmt='o', markersize=4, capsize=3, capthick=1, ecolor='black', color='black')

# plot accuracy score FMA small hoverfit
plt.plot(metricsFMAsmall['n_train_'], metricsFMAsmall['acc_mean_overfit'], markersize=4, color='blue', animated=True,linestyle=':')
plt.errorbar(metricsFMAsmall['n_train_'], metricsFMAsmall['acc_mean_overfit'], yerr=[2*i for i  in metricsFMAsmall['acc_std_overfit']] , fmt='o', markersize=4, capsize=3, capthick=1, ecolor='black', color='black')


# plot accuracy score FMA small handcrafted
plt.plot(metricsFMAsmall_handcrafted['n_train_'], metricsFMAsmall_handcrafted['acc_mean'], markersize=4, label='FMA-small-handcrafted', color = "skyblue")
plt.errorbar(metricsFMAsmall_handcrafted['n_train_'], metricsFMAsmall_handcrafted['acc_mean'], yerr=[2*i for i  in metricsFMAsmall_handcrafted['acc_std']], fmt='o', markersize=4, capsize=3, capthick=1, ecolor='black', color='black')

# plot accuracy score FMA small handcrafted overfit
plt.plot(metricsFMAsmall_handcrafted['n_train_'], metricsFMAsmall_handcrafted['acc_mean_overfit'], markersize=4, animated=True,linestyle=':', color = "skyblue")
plt.errorbar(metricsFMAsmall_handcrafted['n_train_'], metricsFMAsmall_handcrafted['acc_mean_overfit'], yerr=[2*i for i  in metricsFMAsmall_handcrafted['acc_std_overfit']] , fmt='o', markersize=4, capsize=3, capthick=1, ecolor='black', color='black')


# plot accuracy score FMA medium
plt.plot(metricsFMAmedium['n_train_'], metricsFMAmedium['acc_mean'], markersize=4, label='FMA-medium', color='purple')
plt.errorbar(metricsFMAmedium['n_train_'], metricsFMAmedium['acc_mean'], yerr=[2*i for i  in metricsFMAmedium['acc_std']], fmt='o', markersize=4, capsize=3, capthick=1, ecolor='black', color='black')

# plot accuracy score FMA medium
plt.plot(metricsFMAmedium['n_train_'], metricsFMAmedium['acc_mean_overfit'], markersize=4, color='purple', linestyle=':')
plt.errorbar(metricsFMAmedium['n_train_'], metricsFMAmedium['acc_mean_overfit'], yerr=[2*i for i  in metricsFMAmedium['acc_std_overfit']], fmt='o', markersize=4, capsize=3, capthick=1, ecolor='black', color='black')

# plot accuracy score FMA medium handcrafted
plt.plot(metricsFMAmedium_handcrafted['n_train_'], metricsFMAmedium_handcrafted['acc_mean'], markersize=4, label='FMA-medium-hancrafted', color='orchid')
plt.errorbar(metricsFMAmedium_handcrafted['n_train_'], metricsFMAmedium_handcrafted['acc_mean'], yerr=[2*i for i  in metricsFMAmedium_handcrafted['acc_std']], fmt='o', markersize=4, capsize=3, capthick=1, ecolor='black', color='black')

# plot accuracy score FMA medium handcrafted
plt.plot(metricsFMAmedium_handcrafted['n_train_'], metricsFMAmedium_handcrafted['acc_mean_overfit'], markersize=4, color='orchid', linestyle=':')
plt.errorbar(metricsFMAmedium_handcrafted['n_train_'], metricsFMAmedium_handcrafted['acc_mean_overfit'], yerr=[2*i for i  in metricsFMAmedium['acc_std_overfit']], fmt='o', markersize=4, capsize=3, capthick=1, ecolor='black', color='black')


# plot accuracy score FMA large
plt.plot(metricsFMAlarge['n_train_'], metricsFMAlarge['acc_mean'], markersize=4, label='FMA-large', color='green')
plt.errorbar(metricsFMAlarge['n_train_'], metricsFMAlarge['acc_mean'], yerr=[2*i for i  in metricsFMAlarge['acc_std']], fmt='o', markersize=4, capsize=3, capthick=1, ecolor='black', color='black')

# plot accuracy score FMA large
plt.plot(metricsFMAlarge['n_train_'], metricsFMAlarge['acc_mean_overfit'], markersize=4, color='green', linestyle=':')
plt.errorbar(metricsFMAlarge['n_train_'], metricsFMAlarge['acc_mean_overfit'], yerr=[2*i for i  in metricsFMAlarge['acc_std_overfit']], fmt='o', markersize=4, capsize=3, capthick=1, ecolor='black', color='black')


# plot accuracy score FMA large handcrafted
plt.plot(metricsFMAlarge_handcrafted['n_train_'], metricsFMAlarge_handcrafted['acc_mean'], markersize=4, label='FMA-large-handcrafted', color='#808000')
plt.errorbar(metricsFMAlarge_handcrafted['n_train_'], metricsFMAlarge_handcrafted['acc_mean'], yerr=[2*i for i  in metricsFMAlarge_handcrafted['acc_std']], fmt='o', markersize=4, capsize=3, capthick=1, ecolor='black', color='black')

# plot accuracy score FMA large handcrafted
plt.plot(metricsFMAlarge_handcrafted['n_train_'], metricsFMAlarge_handcrafted['acc_mean_overfit'], markersize=4, color='#808000', linestyle=':')
plt.errorbar(metricsFMAlarge_handcrafted['n_train_'], metricsFMAlarge_handcrafted['acc_mean_overfit'], yerr=[2*i for i  in metricsFMAlarge_handcrafted['acc_std_overfit']], fmt='o', markersize=4, capsize=3, capthick=1, ecolor='black', color='black')


# set the range on y-axis
plt.ylim(0,1.1)
# Set x-axis to logarithmic scale
plt.xscale("log")

# set the title
plt.title("Different Datasets Classification Accuracy vs Number of Training Examples")
# set the labels on x-axis and y-axis
plt.xlabel("Number of Traning Examples")
plt.ylabel("Accuracy Score")


# Add a legend with the combined labels
plt.legend() 

# display the background grid
plt.grid(True, linestyle='--', linewidth=0.4, color='gray')
plt.show()
```
